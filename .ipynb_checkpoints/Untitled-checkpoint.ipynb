```python
import os
print(os.listdir(r'C:\Users\admin\Desktop\DL AAT\chest_xray'))
data_dir = r"C:\Users\admin\Desktop\DL AAT\chest_xray"



train_dir = os.path.join(data_dir, 'train')
val_dir   = os.path.join(data_dir, 'val')
test_dir  = os.path.join(data_dir, 'test')




#2.2 Set Up ImageDataGenerators

# Set image size and batch size
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = (150, 150)
BATCH_SIZE = 32

# For training: apply augmentation + rescale
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.2,
    horizontal_flip=True
)

# For validation and test: only rescale
test_val_datagen = ImageDataGenerator(rescale=1./255)

# Load training images
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'  # pneumonia vs normal → binary classification
)

# Load validation images
val_generator = test_val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

# Load test images
test_generator = test_val_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False  # Important for evaluation/prediction
)
```

```python
📊 2.3 Preview Augmented Images
```

```python
# Display 5 sample augmented training images
sample_images, sample_labels = next(train_generator)

plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(sample_images[i])
    plt.title("Pneumonia" if sample_labels[i] == 1 else "Normal")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
✅ STEP 3: Building the CNN Model (Jupyter Notebook)
```

```python
🛠️ 3.2 Build a Simple CNN with Keras
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Input(shape=(150, 150, 3)),               # 👈 add this layer
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])





model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model_architecture.png', show_shapes=True, show_layer_names=True)

```

```python
🔁 4.1 Start Model Training
```

```python
EPOCHS = 10  # You can increase to 15 or 20 for better accuracy

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=EPOCHS,
    validation_data=val_generator,
    validation_steps=val_generator.samples // val_generator.batch_size
)
```

```python
✅ Step 7: Transfer Learning Using VGG16 (in Jupyter Notebook) 
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
```

```python
2️⃣ Load the Pre-trained VGG16 Model (without its classifier head)
```

```python
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))

# Freeze initial layers, unfreeze last 8
for layer in base_model.layers[:-8]:
    layer.trainable = False
    
```

```python
3️⃣ Add Custom Classification Head
```

```python
# Add classifier head
x = base_model.output
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=output)
```

```python
4️⃣ Compile the Model
```

```python

from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=1e-5), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
5️⃣ Load Data Generators Again (if not already)
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.3,
    horizontal_flip=True
)

test_val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir, target_size=(150, 150), batch_size=32, class_mode='binary')
val_generator   = test_val_datagen.flow_from_directory(val_dir, target_size=(150, 150), batch_size=32, class_mode='binary')
```

```python
from sklearn.utils import class_weight
from tensorflow.keras.callbacks import ReduceLROnPlateau

class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}

lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6)
```

```python
6️⃣ Train the Mod
```

```python
model.fit(train_generator,
          validation_data=val_generator,
          epochs=10,
          class_weight=class_weight_dict,
          callbacks=[lr_scheduler])
```

```python
test_loss, test_acc = model.evaluate(test_generator)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
def preprocess_image(path):
    img = tf.keras.utils.load_img(path, target_size=(150, 150))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array
```

```python
✅ Step 3: Define Grad-CAM Function
```

```python
def make_gradcam_heatmap(img_array, model, last_conv_layer_name):
    grad_model = tf.keras.models.Model(
        [model.inputs], 
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, 0]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = np.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()
```

```python
# ✅ 3. Grad-CAM heatmap overlay on the original image
def overlay_gradcam(img_path, heatmap, alpha=0.4):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (150, 150))
    heatmap = cv2.resize(heatmap, (150, 150))
    heatmap = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_TURBO)
    superimposed_img = heatmap_color * 0.2 + img
    return cv2.cvtColor(superimposed_img.astype('uint8'), cv2.COLOR_BGR2RGB)
```

```python
✅ Step 4: Generate Heatmap
```

```python
# Replace 'block5_conv3' with the last conv layer of your model (VGG16)
last_conv_layer_name = 'block5_conv3'

# Generate Grad-CAM heatmap
heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name)
```

```python
✅ Step 5: Display Image with Heatmap Overlay
```

```python
def display_gradcam(original_img_path, heatmap, alpha=0.4):
    img = cv2.imread(original_img_path)
    img = cv2.resize(img, (150, 150))
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    superimposed_img = heatmap_color * alpha + img

    plt.imshow(cv2.cvtColor(superimposed_img.astype('uint8'), cv2.COLOR_BGR2RGB))
    plt.title("Grad-CAM Overlay")
    plt.axis('off')
    plt.show()
```

```python
# 📂 Select test images
import random
normal_dir = "chest_xray/test/NORMAL"
pneumonia_dir = "chest_xray/test/PNEUMONIA"

normal_images = random.sample(os.listdir(normal_dir), 3)
pneumonia_images = random.sample(os.listdir(pneumonia_dir), 3)

# 🧠 Analyze and show each image
for category, images in [("NORMAL", normal_images), ("PNEUMONIA", pneumonia_images)]:
    print(f"\n📂 Category: {category}")
    for file in images:
        path = os.path.join(normal_dir if category == "NORMAL" else pneumonia_dir, file)

        # Preprocess + predict
        img_array = preprocess_image(path)
        prediction = model.predict(img_array)[0][0]
        label = "Pneumonia" if prediction >= 0.5 else "Normal"
        confidence = prediction if prediction >= 0.5 else 1 - prediction

        # Generate Grad-CAM
        heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name="block5_conv3")
        final_image = overlay_gradcam(path, heatmap)

        # Show image + prediction
        plt.imshow(final_image)
        plt.title(f"📸 {file} → {label} ({confidence:.2f})", fontsize=10)
        plt.axis('off')
        plt.show()
```
